```python
## scaling the data
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
d1=['Pregnancies','Outcome']
data1=sc.fit_transform(data.drop(d1,axis=1))
```

```python
data2=pd.DataFrame(data1,columns=['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedif=greeFunction','Age'])
```

```python
final_df=pd.concat([data2,con_data],axis=1)
```

```python
# model creation
#split the data into x&y
x=final_df.drop('Outcome',axis=1)
y=final_df.Outcome
```

```python
# split training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()#obj creation
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_pred_prob=model.predict_proba(x_test)
```

```python
y_pred_prob
```

```python
df=y_test.copy()
```

```python
df['y_pred']=y_pred
```

```python
print(df)
```

# model evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred) # y_test is our actual data
print(cm)
```

```python
recall=recall_score(y_test,y_pred)
recall
```

```python
f1score=f1_score(y_test,y_pred)
f1score
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
from sklearn.metrics import roc_curve, auc
false_postive_rate,true_posive_rate,threshold=roc_curve(y_test,y_pred)
roc_auc_score=auc(false_postive_rate,true_posive_rate)
print(roc_auc_score) #since roc_auc score is near to 1, we say its good model
```
